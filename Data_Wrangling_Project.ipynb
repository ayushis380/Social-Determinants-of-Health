```python
Countydataframe = pd.read_csv("countydataset.csv")
Countydataframe
```

```python
mean_parkdistance = Countydataframe['Park_Distance'].mean()
#print(Countydataframe.groupby(['Park_Distance', 'Walkscore']).mean())
print('mean park distance: ' + str(mean_parkdistance))
mean_walkscore = Countydataframe['Walkscore'].mean()
print('mean walkscore: ' + str(mean_walkscore))
median_parkdistance = Countydataframe['Park_Distance'].median()
print('median park distance: ' + str(median_parkdistance))
median_walkscore = Countydataframe['Walkscore'].median()
print('median walk score: ' + str(median_walkscore))
mode_parkdistance = Countydataframe['Park_Distance'].mode()[0]
print('mode park distance: '+ str(mode_parkdistance))
mode_walkscore = Countydataframe['Walkscore'].mode()[0]
print('mode walk score: '+ str(mode_walkscore))
```

A Walk Score is a rating system that measures the walkability of a location on a scale of 0 to 100, with a higher score indicating that daily errands can be accomplished on foot.
a score of 0.57 should be considered in context and may not always reflect the true walkability of a location.
A score of 0.57 suggests that some amenities, such as grocery stores, restaurants, and parks, may be within walking distance, but not all errands can be accomplished on foot. The exact interpretation of this score may depend on the specific location and what amenities are nearby.

```python
Filled_counties = pd.read_csv("fillcountydataset.csv")
Filled_counties
```

```python
Filled_counties = Filled_counties[['Year','County','Park_Distance','Walkscore']]
Filled_counties
```

```python
Sort_counties = Filled_counties.groupby(['Year','County'], as_index=False).median()
Sort_counties
```

```python
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('fillcountydataset.csv')

# Create a boxplot
plt.boxplot([data['Park_Distance'], data['Walkscore']])
plt.xticks([1, 2], ['Park Distance', 'Walkscore'])
plt.ylabel('Value')
plt.title('Boxplot : Park Distance and Walkscore')
plt.show()
```

```python
df = Sort_counties
# Set the year column as the index
df = df.set_index('Year')

# Get unique county names
counties = df['County'].unique()

# Plot the data for each year
for year in df.index.unique():
    # Create a sub-dataframe for the year
    sub_df = df.loc[year]
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(10,6))
    # Create empty lists for handles and labels for each county
    handles = []
    labels = []
    # Iterate through the county names and plot each column
    for i, county in enumerate(counties):
        # Get the data for the county
        county_data = sub_df.loc[sub_df['County']==county]
        # Create the bar chart with park distance and walk score and save the handles and labels
        park_handle = ax.bar(i*2, county_data['Park_Distance'], color='green', width=0.8, alpha=0.5)
        walk_handle = ax.bar(i*2+0.8, county_data['Walkscore'], color='blue', width=0.8, alpha=0.5)
        handles.append(park_handle)
        handles.append(walk_handle)
    labels.append(f'Park Distance')
    labels.append(f'Walk Score')
    # Set the x-tick labels to be the county names
    ax.set_xticks(range(0,len(counties)*2,2))
    ax.set_xticklabels(counties)
    # Add labels and title
    ax.set_xlabel('County')
    ax.set_ylabel('Value')
    ax.set_title(f'Data for Year {year}')
    # Add the legend with the handles and labels for each county
    ax.legend(handles, labels, loc='upper left')
    # Show the plot
    plt.show()
```
